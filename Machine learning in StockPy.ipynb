


# Data Reading
Here First we're going to read the data, first we need all the names of the firms for which we are going to predict the target variables.
         
'1COMPANY INFO.xlsx' has this information in the data



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
company_details = pd.read_excel('data/1COMPANY INFO.xlsx')
company_details.head()
```

# Companies

```python
df  = pd.DataFrame(np.zeros((821,12)))
col = ['Openprice','Highprice','Lowprice'
       ,'Closeprice','minlowprice','maxhighprice',
       'HighminusOpen','CloseminusLow','CloseminusOpen','max_variance','Volume','count_entries']
df.columns = col
df.index = company_details['File Name']
```

files that have data of more than two years before 2017
-A1M-Price
-AAP_Price

# Feature Generation from data

```python
not_avail = ['ACCT','ADDV','ALELIO','ATRLJ','CATE','CLDR','NTGR','NOKIA','RNBS','FRIM','SAS','SPILTAN']
for company_name in company_details['File Name']:
        if company_name not in not_avail:
            temp = pd.read_excel('data/'+company_name+'-price.xls')
            temp['year']=pd.DatetimeIndex(temp['Date']).year
            col = ['Openprice','Highprice','Lowprice','Closeprice','HighminusOpen','CloseminusLow','CloseminusOpen','Volume']
            if (temp[temp.year==2017].shape[0]>0):
                temp['HighminusOpen'] = temp.Highprice  - temp.Openprice
                temp['CloseminusLow'] = temp.Closeprice - temp.Lowprice
                temp['CloseminusOpen'] = temp.Closeprice - temp.Openprice
                for feature_col in col:
                    df.loc[company_name][feature_col] = temp[feature_col].mean()
                    df.loc[company_name]['minlowprice'] = temp.Lowprice.min()
                    df.loc[company_name]['maxhighprice'] = temp.Lowprice.min()
                    df.loc[company_name]['max_variance'] = temp.CloseminusOpen.max()
                    df.loc[company_name]['count_entries'] = temp.Openprice.count()
```
